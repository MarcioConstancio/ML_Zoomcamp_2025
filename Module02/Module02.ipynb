## 2.5 [Linear Regression](https://youtu.be/Dn1eTQLsOdA?si=g42pm9qnwIfP4Qe_)



```python
df_train
```

```python
df_train.iloc[10], y_train[10]
#df_train.select_dtypes(exclude='object')
```

```python
xi = [453,11,86]
w0 =7.17
w = [0.01,0.04,0.002]

#g(xi) = 7.17 + 453*0.01 + 11*0.04 + 86*0.002

def linear_regression(xi):
    
    n = len(xi)
    
    pred = w0
    
    for j in range(n):
        pred = pred + w[j]*xi[j]
    return pred

    
```

```python
np.expm1(w0)
np.expm1(linear_regression(xi))
```
